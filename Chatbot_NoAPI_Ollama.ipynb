```python
from langchain.chat_models import ChatOllama
from langchain.schema import SystemMessage, HumanMessage, AIMessage

# Load the local model using Ollama
llm = ChatOllama(model="mistral")  # or 'llama3', 'gemma', etc.

def get_completion_from_messages(messages, temperature=0.7):
    # Convert your messages to LangChain format
    langchain_messages = []
    for msg in messages:
        role = msg["role"]
        content = msg["content"]
        if role == "system":
            langchain_messages.append(SystemMessage(content=content))
        elif role == "user":
            langchain_messages.append(HumanMessage(content=content))
        elif role == "assistant":
            langchain_messages.append(AIMessage(content=content))
    
    # Run the model with temperature
    return llm(langchain_messages, temperature=temperature)
```

```python
messages =  [  
{'role':'system', 'content':'You are an assistant that speaks like Shakespeare.'},    
{'role':'user', 'content':'tell me a joke'},   
{'role':'assistant', 'content':'Why did the chicken cross the road'},   
{'role':'user', 'content':'I don\'t know'}  ]

response = get_completion_from_messages(messages, temperature=1)
print(response)

```

```python
# OrderBot

def collect_messages(_):
    prompt = inp.value_input
    inp.value = ''
    context.append({'role': 'user', 'content': prompt})

    # Call the LLM and extract .content if the response is an AIMessage
    response = get_completion_from_messages(context)
    if hasattr(response, "content"):
        response = response.content

    context.append({'role': 'assistant', 'content': response})

    panels.append(pn.Row('User:', pn.pane.Markdown(prompt, width=600)))
    panels.append(pn.Row('Assistant:', 
                         pn.pane.HTML(f"<div style='background-color:#F6F6F6;padding:10px'>{response}</div>", 
                                      width=600)))
    
    return pn.Column(*panels)
```
